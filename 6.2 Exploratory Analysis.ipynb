# 06. Categorical Plots

## GDP per capita vs overall rank

```python
sns.histplot(df_2019_prepared_sub['GDP per capita'], bins = 20, kde = True)
```

```python
df_2019_prepared_sub.loc[df_2019_prepared_sub['GDP per capita'] < 0.75, 'GDP per capita category'] = 'Low GDP per capita'
```

```python
df_2019_prepared_sub.loc[(df_2019_prepared_sub['GDP per capita'] > 0.75) & (df_2019_prepared_sub['GDP per capita'] < 1.25), 'GDP per capita category'] = 'Medium GDP per capita'
```

```python
df_2019_prepared_sub.loc[df_2019_prepared_sub['GDP per capita'] > 1.25, 'GDP per capita category'] = 'High GDP per capita'
```

```python
df_2019_prepared_sub['GDP per capita category'].value_counts(dropna = False)
```

```python
df_2019_prepared_sub.columns
```

```python
sns.set(style = "ticks")

g = sns.catplot(x="GDP per capita", y="Overall rank", hue="GDP per capita category", data=df_2019_prepared_sub)
```
